Import Libraries

```python
import pandas as pd
```

Read the file

```python
df = pd.read_csv('Customer1 - Customer1.csv')
display(df)
```

Missing Values

```python
# check how many missing values in tha columns
print(df.isnull().sum())
```

```python
# replace the null values in the gender and perfix columns
df['FullName'] = df['Prefix'].fillna('') + ' ' + df['FirstName'].fillna('') + ' ' + df['LastName'].fillna('')
df['FullName'] = df['FullName'].str.strip()
df['Gender'].fillna('Unknown', inplace=True)
print(df.isnull().sum())
```

Know the type of each column

```python
print(df.dtypes)
```

Changing AnnualIncome and BirthDate  data Types

```python
#AnnualIncom from object to float
df['AnnualIncome'] = df['AnnualIncome'].astype(str).str.replace('$','').str.replace(',','').astype(float)

#BirthDate from object to date
df['BirthDate'] = pd.to_datetime(df['BirthDate'])

print(df.dtypes)
display(df)
```

Check is there any  Duplicates


```python
duplicate_rows = df.duplicated().sum()
print({duplicate_rows})
```
